# Selecting A Suitable Place to Live in North York, ON

## A Coursera Capstone Project Assignment

### Data

In order to help interested people to choose a North York neighbourhood for settling permanently, location data for the city of Toronto and nearby areas will be used from the following Wikipedia website:

https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M

In order to get hold of the geographical longitudinal and latitudinal coordinates, the following website will be consulted to access the ‘Toronto_long_lat_data.csv’ file. The coordinates will help in developing maps of the area, as well as to analyze the neighbourhoods.

http://cocl.us/Geospatial_data

In order to get information on the nearby venues of the selected areas, Foursquare website will be used. Location data, along with information on nearby venues and points of interest can be obtained from Foursquare website, which is a location data provider. An appropriate url will be used to create the API request for the desired information.

Housing price data for the neighbourhoods in the North York region will be procured from the following website. The website contains the data related to average house price for a given postal code in Canada:

https://housepricehub.com/areas

Data collected from the aforementioned Wikipedia page will be merged with the geographical coordinates data which will be obtained from the Geospatial data website. This will enable us to create maps and analyze different neighbourhoods. Aan API request will be made using the Foursquare website to obtain data on nearby venues for the areas of interest. This will enable us to divide different North York neighbourhoods into clusters based on venues around them. It will also help a person make a decision to choose between various neighbourhoods based on points of interest alone. To add to this, the house price data will be used to decide the best neighbourhood to live within a cluster of similar neighbourhoods. The clustering analysis with be carried out using the K-means technique. Analysis of different clusters along with the house price data will help a person choose the best neighbourhood in the region of North York, Toronto, ON.
