# Selecting A Suitable Place to Live in North York, ON

## A Coursera Capstone Project Assignment 

### Introduction/Business Problem

Canada has always been an attractive place for immigrants and workers from around the globe. Greater Toronto Area (GTA) is one of the major areas in Canada which attracts a lot of people who come to Canada. People arriving at Canada carry a variety of preferences with regards to the location where they would like to settle in a given city or area. They also belong to a variety of diverse backgrounds: socially, culturally and financially. Different people have different preferences about selecting the location where they would like to live permanently. This primarily depends on the surrounding locations, which includes availability of transportation facilities, hospitals, schools, parks, restaurants belonging to a particular cuisine, places of other recreational activities, etc. Many neighbourhoods in a given area may be similar in terms of the abovementioned factors. House prices can be taken into account while deciding to choose between two neighbourhoods which are similar.

North York is one of the six administrative districts in Toronto, ON, Canada, which was created as a township in 1922, and was included in Metropolitan Toronto in 1954. Currently, the house prices in North York are higher than other regions in the GTA (such as Brampton), however, there are people who can afford owning a house here and who would prefer to live in the region, given it’s close proximity to Downtown Toronto and other nearby locations. Therefore, this project was designed to help people select a neighbourhood in North York, based on their preferences, common venues in the area, and average house prices.
